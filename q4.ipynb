```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im = cv.imread('images/woman_in_shadow.avif', cv.IMREAD_GRAYSCALE)
assert im is not None, 'Image not found'

# (a) 
ret, mask = cv.threshold(im, 0, 255, cv.THRESH_BINARY_INV + cv.THRESH_OTSU)

print(f"The resulting Otsu threshold value is: {ret}")

# (b) 
img_equalized = im.copy()

# Apply equalization to the masked region
foreground_pixels = im[mask == 255]
equalized_foreground = cv.equalizeHist(foreground_pixels)
img_equalized[mask == 255] = equalized_foreground.flatten()

fig, ax = plt.subplots(1, 3, figsize=(12, 6))

ax[0].imshow(im, cmap='gray')
ax[0].set_title('Original Grayscale')
ax[0].axis('off')

ax[1].imshow(mask, cmap='gray')
ax[1].set_title('Otsu Binary Mask')
ax[1].axis('off')

ax[2].imshow(img_equalized, cmap='gray')
ax[2].set_title('Equalized Foreground')
ax[2].axis('off')

plt.show()
```
