```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

im = cv2.imread('images/a1images/emma_salt_and_pepper_noice.jpg', cv2.IMREAD_GRAYSCALE)
assert im is not None, 'Image not found'
    
# Apply Gaussian Smoothing
gaussian_blur = cv2.GaussianBlur(im, (5, 5), 0)

# Apply Median Filtering
median_blur = cv2.medianBlur(im, 5)

fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(im, cmap='gray', vmin=0, vmax=255)
ax[0].set_title('Original')
ax[0].axis('off')

ax[1].imshow(gaussian_blur, cmap='gray', vmin=0, vmax=255)
ax[1].set_title('Gaussian Smoothing')
ax[1].axis('off')

ax[2].imshow(median_blur, cmap='gray', vmin=0, vmax=255)
ax[2].set_title('Median Smoothing')
ax[2].axis('off')

plt.show()
```
